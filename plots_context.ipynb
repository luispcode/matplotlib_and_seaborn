```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y= np.sin(x)
```

#### metodo orientado a objetos.

```python
fig, axes = plt.subplots(1,2,figsize=(12,8)) #aca se define el numero de graficos, (una fila 2 columnas) figsize tamaño del lienzo.
axes[0].plot(x,y, label="$sin(x)$") #metodo para graficar.
axes[0].set_title('Relcion X - Y') #colocar titulo.
axes[0].set_xlabel('X') #colocar etiqueta eje x
axes[0].set_xlabel('Y') #colocar etiqueta eje y
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relcion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[0].legend()
```

#### metodo sin orientado a objetos.

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right',bbox_to_anchor=(1,0.8))
```
