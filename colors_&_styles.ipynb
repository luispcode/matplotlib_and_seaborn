```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
```

#### estilos disponibles

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background') #set theme.
```

#### estilos de linea como en matlab eg:('r--')

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

#### pyplot style
--alpha es transparencia

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color = 'green', alpha=0.8, linewidth=6, linestyle='-', marker='o', markersize=8, markerfacecolor='#CCFFCC')
ax.plot(x,x+2,color = 'blue', linewidth=2, linestyle='--', marker='X')
ax.plot(x,x+3,color = '#56FF56', linewidth=8, linestyle='dashed', marker='v')
ax.plot(x,x+4,color = '#CCFFCC', linewidth=4, linestyle=':', marker='P')
```
