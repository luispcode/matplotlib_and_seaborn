```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y, 'b')
axes[1].plot(y,x, 'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
```

```python
fig, axes = plt.subplots(2,4)#2 filas 4 columnas en el lienzo.
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
axes[1,0].plot(x,y,'ro')
axes[1,1].plot(y*2,x/2,'gx')
axes[1,2].plot(x+10,y*3,'salmon')
axes[1,3].plot(y/2,x*-1)

fig.tight_layout() #separar graficos.
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
ax5.plot(x,y,'ro')
ax6.plot(y*2,x/2,'gx')
ax7.plot(x+10,y*3,'salmon')
ax8.plot(y/2,x*-1)

fig.tight_layout() #separar graficos.
```
