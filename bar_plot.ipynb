```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### bar plot o tabla para variables categoricas y analisis univariable

```python
country = ['INDIA', 'JAPAN', 'MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
data = np.random.randint(1,50,100)
data
```

```python
plt.bar(country, population, width=0.5, color=['green', 'blue'])
plt.xticks(np.arange(5), ('India', 'Japon','Mexico','Colombia','Alemania'),rotation=45)
plt.show()
#colocar contexto.
```

#### Histograma 
--distribucion de los datos, utilizando bins

```python
plt.hist(data, bins=10, width=5,histtype='bar') #bins=10 by default
plt.show()
```

```python
plt.hist(data, bins=10,histtype='step') #bins=10 by default
plt.show()
```

#### boxplot

```python
data = np.append(data,200)
plt.boxplot(data, vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

```python
plt.boxplot(data, vert=True,patch_artist=True,notch=False)
plt.show()
```

#### Scatter plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30* np.random.rand(N)) **2
colors = np.random.rand(N)
```
