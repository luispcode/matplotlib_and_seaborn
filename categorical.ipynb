```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
sns.countplot(data=tips,x='day',hue='sex') #hue segmentar por.
plt.show()
```

```python
plt.figure(figsize=(6,7))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex', dodge=True) #hue segmentar por.
plt.show()
```

```python
plt.figure(figsize=(6,7))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex', dodge=True) #hue segmentar por.
plt.show()
```

```python
plt.figure(figsize=(7,8))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex', dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex', dodge=True, color='0', marker='o') #hue segmentar por.
plt.show()
```

```python
plt.figure(figsize=(7,8))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex', dodge=True, split=True, palette='bright')
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
```

#### cambiar e iterar varios graficos con el metodo catplot
-- cambiando el kind se pueden probar varios graficos

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex', dodge=True, palette='bright',kind='box', col='time')
plt.show()
```
